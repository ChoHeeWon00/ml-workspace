```python
import pandas as pd
```

```python
df = pd.read_csv("../../data_set/2.데이터 클린징/loan_train_quiz_cleaning.csv")
df.head(2)
```

```python
df.columns
```

```python
features = ['Gender', 'Married', 'Dependents', 'Education',
       'Self_Employed', 'ApplicantIncome', 'CoapplicantIncome', 'LoanAmount',
       'Loan_Amount_Term', 'Credit_History', 'Property_Area']

label = 'Loan_Status'
```

```python
X, y = df[features],df[label]
```

```python
df.shape
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2 )
```

```python
X_train.shape, X_test.shape, len(y_train), len(y_test)
```

### 모델 생성 및 학습
* 모델 생성 : 변수 = DecisionTreeClassifier()
* 학습 : 변수.fit(X_train, y_train)
    - 문제와 정답을 알고리즘에 학습
* 예측 : 변수.predict(X_test)
    - 확인용 문제를 predict에 넣으면 정답을 알려준다

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dtc = DecisionTreeClassifier() # 모델 생성
dtc.fit(X_train, y_train) #학습
```

```python
pred = dtc.predict( X_test ) # 20% 학습하지 않은 데이터. 예상 답안 뽑음
pred
```

```python
print(y_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test, pred)
```
